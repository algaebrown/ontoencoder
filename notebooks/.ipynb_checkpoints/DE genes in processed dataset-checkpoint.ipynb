```python
df = pd.read_pickle(root_dir + data_id['GEO accession'][1]+'.pickle')
```

```python
# parsing cell type
cleaned_name = [cell[:-1] if cell[-1] == ' ' else cell for cell in df.columns ]
celltype = [cell.split(' ')[-1] for cell in cleaned_name]
species = ['mice' if 'Mouse' in cell else 'human' for cell in cleaned_name]
age = [int(cell.split('yo')[0]) if 'yo' in cell else 0 for cell in cleaned_name]
```

```python
obs = pd.DataFrame(columns = cleaned_name)
obs.loc['cell type'] = celltype
obs.loc['species'] = species
obs.loc['age'] = age
```

```python
obs
```

```python
from anndata import AnnData
adata = AnnData(X = df.transpose())
adata.obs = obs.transpose()
adata
```

```python
# add the total counts per cell as observations-annotation to adata
adata.obs['total_expression'] = adata.X.sum(axis=1)
```

```python
adata.obs['total_expression'].sort_values().plot(title = 'sorted total expression per cell')
# don't know what kind of normalization they use? FPKM maybe (confirmed)
```

```python
adata.var['total_expression'] = adata.X.sum(axis=0)
adata.var['total_expression'].sort_values().plot(title = 'sorted total expression per gene')
```

```python
sc.pl.highest_expr_genes(adata, n_top=20, show = True) # take a look at highest expressed genes
# Computes, for each gene, the fraction of counts assigned to that gene within a cell. The n_top genes with the highest mean fraction over all cells are plotted as boxplots.
```

```python
sc.pp.log1p(adata) # FPKM is not logged
adata.raw = adata
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
```

```python
# what are the highly variable gene that are used to cluster
adata.var.loc[adata.var['highly_variable']]
```

```python
sc.pl.highly_variable_genes(adata)
```

```python
adata.raw = adata
adata = adata[:, adata.var.highly_variable]
```

```python
sc.pp.scale(adata, max_value=10) # scale to unit variance
```

```python
adata.obs
```

```python
sc.tl.pca(adata, n_comps = 20)
sc.pl.pca(adata, color=['age','species', 'cell type'])
# apparently mouse and human transcriptome are different. Fetal and adult are different too.
```

```python
sc.pl.pca_variance_ratio(adata, log=True)
```

```python
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=20)
sc.tl.umap(adata)
```

```python
sc.pl.umap(adata, color = ['age', 'cell type', 'species'])
```
